```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("NFHS_5_India_Districts_Factsheet_Data.xlsx")

print("Shape:", df.shape)
df.head()
```

```python
cols = [
    "State/UT",
    "District Names",

    "Women age 15 years and above wih high or very high (>140 mg/dl) Blood sugar level or taking medicine to control blood sugar level23 (%)",
    "Men age 15 years and above wih high or very high (>140 mg/dl) Blood sugar level  or taking medicine to control blood sugar level23 (%)",

    "Women age 15 years and above wih Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)",
    "Men age 15 years and above wih Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)",

    "Women (age 15-49 years) who are overweight or obese (BMI ≥25.0 kg/m2)21 (%)",

    "Women age 15 years and above who use any kind of tobacco (%)",
    "Men age 15 years and above who use any kind of tobacco (%)",

    "Women age 15 years and above who consume alcohol (%)",
    "Men age 15 years and above who consume alcohol (%)",


    "All women age 15-49 years who are anaemic22 (%)",

    "Households with any usual member covered under a health insurance/financing scheme (%)"
]

df = df[cols]
```

```python
df.columns = [
    "state",
    "district",
    "diabetes_women",
    "diabetes_men",
    "bp_women",
    "bp_men",
    "obesity_women",
    "tobacco_women",
    "tobacco_men",
    "alcohol_women",
    "alcohol_men",
    "anaemia_women",
    "insurance_coverage"
]
```

```python
for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["diabetes_avg"] = df[["diabetes_women", "diabetes_men"]].mean(axis=1)
df["bp_avg"] = df[["bp_women", "bp_men"]].mean(axis=1)
df["tobacco_avg"] = df[["tobacco_women", "tobacco_men"]].mean(axis=1)
df["alcohol_avg"] = df[["alcohol_women", "alcohol_men"]].mean(axis=1)
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
d_mean = df["diabetes_avg"].mean()
bp_mean = df["bp_avg"].mean()

def risk_label(row):
    if row["diabetes_avg"] > d_mean and row["bp_avg"] > bp_mean:
        return "High"
    elif row["diabetes_avg"] > d_mean or row["bp_avg"] > bp_mean:
        return "Medium"
    else:
        return "Low"

df["health_risk_level"] = df.apply(risk_label, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["risk_encoded"] = le.fit_transform(df["health_risk_level"])
```

```python
final_df = df[
    [
        "state",
        "district",
        "diabetes_avg",
        "bp_avg",
        "obesity_women",
        "tobacco_avg",
        "alcohol_avg",
        "anaemia_women",
        "insurance_coverage",
        "health_risk_level",
        "risk_encoded"
    ]
]

final_df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

gender_df = pd.DataFrame({
    "Women": [
        df["diabetes_women"].mean(),
        df["bp_women"].mean()
    ],
    "Men": [
        df["diabetes_men"].mean(),
        df["bp_men"].mean()
    ]
}, index=["Diabetes (15+)", "High BP (15+)"])

gender_df.plot(kind="bar", figsize=(7,5))
plt.title("Gender-wise Comparison of Diabetes and Blood Pressure")
plt.ylabel("Average Percentage")
plt.xlabel("Health Indicator")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()

```

```python
import seaborn as sns

corr_cols = [
    "diabetes_avg",
    "bp_avg",
    "obesity_women",
    "tobacco_avg",
    "alcohol_avg",
    "anaemia_women",
    "insurance_coverage"
]

corr_matrix = final_df[corr_cols].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Health Indicators")
plt.show()

```

```python
disease_avg = {
    "Diabetes": final_df["diabetes_avg"].mean(),
    "High Blood Pressure": final_df["bp_avg"].mean(),
    "Obesity (Women)": final_df["obesity_women"].mean(),
    "Anaemia (Women)": final_df["anaemia_women"].mean(),
    "Tobacco Use": final_df["tobacco_avg"].mean(),
    "Alcohol Consumption": final_df["alcohol_avg"].mean()
}

disease_df = pd.DataFrame(
    disease_avg.items(),
    columns=["Disease", "Average Percentage"]
)

plt.figure(figsize=(9,5))
sns.barplot(data=disease_df, x="Disease", y="Average Percentage")
plt.title("Disease Comparison Across Population")
plt.xticks(rotation=30)
plt.grid(axis="y")
plt.show()
```
